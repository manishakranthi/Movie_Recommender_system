```python
import ast
```

```python
def convert(text, key='name', limit=3):
    L = []
    counter = 0
    for item in ast.literal_eval(text):
        if counter < limit:
            L.append(item.get(key))
            counter += 1
    return L
```

```python
credits['cast'] = credits['cast'].apply(convert)
```

```python
credits.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
credits['crew'] = credits['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:str(x).split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
df = movies.drop(columns=['overview','genres','keywords','cast','crew','imdb_id'])
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
df[df['title'] == 'Grumpier Old Men']
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
        
```

```python
recommend('Grumpier Old Men')
```

```python
recommend('Romance')
```
